```python
import pandas as pd
import matplotlib.pyplot as plt

red_wine_data = pd.read_csv('winequality-red.csv')
red_wine_data[['fixed acidity',"volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]]= red_wine_data['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'].str.split(';',expand=True)
red_wine_data=red_wine_data.drop(columns='fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"')
display(red_wine_data.describe())
red_wine_quality = red_wine_data['quality']
red_wine_data = red_wine_data.drop(columns='quality')
display(red_wine_data.head(10))

class_distribution = red_wine_quality.value_counts()
plt.bar(class_distribution.index, class_distribution)
plt.ylabel("Frequency")
plt.show()
```
